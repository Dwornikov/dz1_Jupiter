5.3 Выберите дом клиенту
Клиент хочет дом без подвала, с двумя этажами, стоимостью до 150000. Какая оценка по состоянию у таких домов в среднем?

### Задание 5.1

```python
df.view.value_counts()
```

```python
num_house=df[(df["waterfront"]==1) & (df["bedrooms"]>=3) & (df["sqft_basement"]>0)]
```

```python
print(f"кол-во домов по первому условию {num_house.shape[0]}")
```

### Задание 5.2

```python
filtr1=df[ (df["yr_built"]>1980) & (df["grade"]>5) & ((df["view"]>0) | (df["waterfront"]>0))]["price"].min()
filtr2=df[ (df["yr_built"]>1980) & (df["grade"]>5) & ((df["view"]>0) | (df["waterfront"]>0))]["price"].max()
```

```python
print(f"Цена на задание 5,2 будет от {filtr1} До {filtr2}")
```

### Задание 5,3

```python
fltr5_3=df[(df["floors"]>=2) & (df["sqft_basement"]==0) & (df["price"]<150000)]["grade"].mean()
```

```python
print(f"Средняя оценка на такие дома {fltr5_3}")
```
